```python
import numpy as np
```

```python
xarr = np.arange(15)
yarr = np.arange(15,30)
```

```python
print(xarr)
print(yarr)
```

```python
cond = np.array([True, False,  True,  False, True, False,  True,  False,True, False,  True,  False, True, False,  True,])
```

```python
result = np.where(cond,xarr,yarr)
result
```
